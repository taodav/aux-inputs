```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

from definitions import ROOT_DIR
from unc.utils import load_info
```

```python
rpg_path = Path(ROOT_DIR, 'results', 'rpg', '213b0798f871ed653e9e6920ed87d21c_Wed Sep 29 14:39:09 2021.npy')
info = load_info(rpg_path)
```

```python
steps_per_diff = 1000
index_to_check = 4
action_name = "SAMPLE"
gb_colors = ["green", "red"]
labels = ['Good', 'Bad']
```

```python
plots = [info['good_diffs'][:, index_to_check], info['bad_diffs'][:, index_to_check]]
fig, ax = plt.figure(), plt.axes()

for label, color, to_plot in zip(labels, gb_colors, plots):
    ax.plot(np.arange(to_plot.shape[0]) * steps_per_diff, to_plot, label=label, color=color)
    
ax.set_xlabel('Environment steps')
ax.set_ylabel(f"Average Q-value difference \nbefore and after checking", rotation=0, labelpad=55)
ax.set_title(f'Average (over rocks) Q-value differences for action {action_name}')
plt.legend(bbox_to_anchor=(1.00, 1), loc='upper left')
```
